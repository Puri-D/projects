```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def remove_stop_words(text):
    words = text.split()
    words = [word for word in words if word not in stop_words]
    words = [word for word in words if word.isalpha()]  # Filter out non-alphabetic words
    return " ".join(words)

udm['published_timestamp'] = pd.to_datetime(udm['published_timestamp'])
udm['Free/Paid'] = udm['Free/Paid'].apply(lambda x: True if x == 'Paid' else False)
udm.dropna(inplace=True)
udm.drop(columns='url')
lower_list = ['course_title', 'subject', 'level']

for i in lower_list:
    udm[i] = udm[i].str.lower()
    udm[i] = udm[i].str.replace('"','')
    #udm[i] = udm[i].apply(remove_stop_words)
```
